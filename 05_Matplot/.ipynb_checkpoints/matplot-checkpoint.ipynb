```python
import matplotlib.pyplot as plt
```

```python
## allow us to see plots inside jupyter notebook
%matplotlib inline

###otherwise : plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

## Functional method

```python
plt.plot(x,y)
plt.xlabel('X Label')       # Set X label
plt.ylabel('Y label')       # set Y label
plt.title('Title')          # set Title
```

```python
### subplot divides parts based on row and column wise
### agr1: num of row, arg2: num of column, arg3: position
plt.subplot(1,2,1)     ## here, nrow = 1, ncol = 2, pos = 1
plt.plot(x,y,'r')      ## 'r' means color red

plt.subplot(1,2,2)     ## here, nrow = 1, ncol = 2, pos = 2
plt.plot(y,x,'b')      ## 'b' means color blue
```

## Object Oriented way

```python
## created a figure object
fig = plt.figure()

## set axes: [left, bottom, width, height]
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.3, 0.3])

axes1.plot(x,y)
axes2.plot(x,y)
```

#### (1) create Subplot

```python
fig, axes = plt.subplots(nrows=3,ncols=3)

## fixed the overlap between plots
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

## we can iterate through exes
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

## we can use index for axes
axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_xlabel('X label')
axes[0].set_ylabel('Y Label')
axes[0].set_title('Set Title')

axes[1].set_xlabel('X label')
axes[1].set_ylabel('Y Label')
axes[1].set_title('Set Title')

plt.tight_layout() ## always put at the end. solves overlaping
```

### (2) Figure size, DPI

```python
## figsize(width, height)
fig = plt.figure(figsize = (3,2), dpi=600) 

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
## figsize(width, height)
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(6,2), dpi = 600)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()

```

### (3) Save figure

```python
fig.savefig('my_picture.png', dpi = 600)
```

## Example

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc = 0)  ## loc = 0 means best location (range 0-10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')
ax.legend(loc = (0.2,0.5))   ## manually set location
```

## (4) Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue',
       linewidth=5, alpha = 0.2,
       linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'blue',
       linewidth=5, alpha = 0.2,
       linestyle='-.')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])


## alpha = transparency; marker = mark points
ax.plot(x,y, color = 'blue',
       linewidth=5, alpha = 1,
       linestyle='-', marker='o', markersize = 20,
       markerfacecolor = 'red', markeredgewidth=3,
       markeredgecolor='black')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])


## alpha = transparency; marker = mark points
ax.plot(x,y, color = 'blue',
       linewidth=5, alpha = 1,
       linestyle='-', marker='o', markersize = 20,
       markerfacecolor = 'red', markeredgewidth=3,
       markeredgecolor='black')


## set x and y limits
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

## Histogram

```python
from random import sample

data = sample(range(1,1000), 100)

plt.hist(data)
```

## Boxplot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

plt.boxplot(data, vert = True, patch_artist = True)
```
